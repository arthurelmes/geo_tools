```python
import pandas as pd
import matplotlib.pyplot as plt
import os

os.chdir('/home/arthur/Dropbox/projects/rr_valley/')

csv = pd.read_csv('rr_valley_brdf_principal_plane.csv')
csv = csv[['modis_band', 'directional_ref', 'vzen']]
print(type(csv))
```

```python
fig, ax = plt.subplots()
fig.patch.set_facecolor('white')

for b in range(1, 8):
    band_name = ' Band ' + str(b) + ' '
    csv_band = csv.loc[csv['modis_band'] == band_name]
    plt.plot(csv_band['vzen'], csv_band['directional_ref'], label=band_name)

ax.set_ylim(0.0, 0.6)
plt.legend(loc='lower left', ncol=2)
ax.set_xlabel('View Zenith Angle')
ax.set_ylabel('Directional Reflectance')
ax.set_title('Principal Plane')
```

```python
# Here, plot just bands 1 and 2 as a scatterplot

fig, ax = plt.subplots(figsize=(10,10))
fig.patch.set_facecolor('white')


csv_b1 = pd.DataFrame(csv[csv['modis_band'] == " Band 1 "])
csv_b2 = pd.DataFrame(csv[csv['modis_band'] == " Band 2 "])
csv_b12 = csv_b1.merge(csv_b2, how='left', on="vzen")

print(csv_b12.head)
plt.plot(csv_b12['directional_ref_x'], csv_b12['directional_ref_y'])
ax.set_ylim(0.3, 0.6)
ax.set_xlim(0.3, 0.6)
ax.set_xlabel('Band 1 Directional Reflectance')
ax.set_ylabel('Band 2 Directional Reflectance')
ax.set_title('Band 1 vs Band 2')
```
